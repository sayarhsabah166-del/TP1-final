````python
for token in tokens:
    t=tokenizer.decode_single_token_bytes(token).decode('utf-8')
    print(t)
````

````python
def tokens_count(prompt:str, model:str="o200k_base"):
    tokenizer = tiktoken.get_encoding(model)
    return len(tokenizer.encode(prompt))
````

````python
print(tokens_count(prompt="Vous êtes un assistant de traduction. "))
````
